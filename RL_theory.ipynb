DIVE IN REGRESSÃO LOGISTICA

O que é Regressão Logística?

A Regressão Logística (RL) é um modelo de machine learning supervisionado de classificação binária. O objetivo é encontrar a relação um alvo categórico (target) e minhas variáveis exploratórias (features). Apesar da designação de regressão, a RL tem por output uma variável categórica binária (ex.: 0/1, True/False, Sim/Não, Feminino/Masculino), que é escolhido através da probabilidade de uma observação ocorrer, se essa probabilidade atingir um limite mínimo desejado, classificamos o evento como 1, caso contrário, 0. A equação que molda o ajuste do modelo é dada por:

Função Sigmóide:
$$f(x) = \frac{1}{1+e^{-\omega x +\beta}}$$

Desse modo se traçarmos nosso limite em 50%, então:
- Se $f_\omega(x) \geq 0.5$, a previsão é a classe 1.
- Se $f_\omega(x) < 0.5$, a previsão é a classe 0.

Como o modelo de RL funciona?

Utilizar o Erro Médio Quadrado (MSE) nessa abordagem geraria uma superfície não convexa, dificultando muito o processo de aprendizado do modelo. Pela configuração da função em uso, a equação de perda (LF) nesse caso é a Entropia Cruzada Binária (Log Loss), designada por:

$$J(w,b) = \frac{1}{N}\sum_{i=1}^{N}[y_i log(\hat{y}_i)+(1-y_i)log(1-\hat{y}_i)]$$

Onde:
$y_i$: é o rótulo real (0 ou 1).
$\hat{y}_i$: é a probabilidade prevista pelo modelo.

Ta mas de onde essa equação de perda surgiu?

Vamos voltar alguns passos, e introduzir uma métrica, 'log-odds', que nada mais é do que a chance de um evento ocorrer, sob a chance de não ocorrer:

Log-Odds:
$$odds = \frac{P(evento)}{1-P(evento)}$$

Outro conceito importante na probabilidade é a probabilidade condicional, qual a chance de A acontecer, dado que B aconteceu. Essa estrutura é defininda por:

Probabilidade Condicionada:
$$Pr(y=1|X) = p(X)$$

onde calculamos a probabilidade da classe 1 ocorrer dado que o evento X ocorreu. Essa função $p(X)$ é dada pela Função Sigmóide apresentada anteriormente.

$$p(X) = \frac{1}{1+e^{-\omega x +\beta}}$$

reorganizando, podemos escrever:

$$\ln\left(\frac{p(X)}{1-p(X)}\right) = \omega x + \beta$$

para chegarmos na LF, vamos usar a abordagem de Máxima Verossimilhança (MLE). Para um evento binário ocorrer, utilizamos a distribuição de Bernulli, onde a probabilidade de um evento y ocorrer é:

$$P(y|x) = \hat{y}^y (1-\hat{y})^{(1-y)}$$

onde:
- $\hat{y}$ é a probabilidade do evento ocorrer
- $y$ ocorrência do evento
- $1-\hat{y}$ probabilidade do evento não ocorrer
- $1-y$ não ocorrência do evento

em todo nosso dataset, o objetivo é maximizar o produto dessas probabilidades (Verossimilhança). Para transformar esse produto sequencial em uma soma sequencial, aplicamos o logaritmo e invertemos o sinal, passando de uma maximização para minimização, e assim chegamos na Log Loss:

$$J(w,b) = -\frac{1}{N} \sum_{i=1}^{N} \underbrace{\left[ y_i \ln(\hat{y}_i) + (1-y_i) \ln(1-\hat{y}_i) \right]}_{\text{Entropia Cruzada Binária}}$$

de forma que:
- Se $y=1$: O custo é $-\ln(\hat{y})$. Se $\hat{y} \to 1$, o custo é $0$. Se $\hat{y} \to 0$, o custo tende ao infinito.
- Se $y=0$: O custo é $-\ln(1-\hat{y})$. Se $\hat{y} \to 0$, o custo é $0$. Se $\hat{y} \to 1$, o custo tende ao infinito.


Mas como o RL aprende?

Assim como na Regressão Linear, na RL existe um processo de otimização da atualização dos valores de $\omega$ e $\beta$ através do Gradiente Descendente. Que assim como antes é a derivada em função dos termos $\omega$ e $\beta$ mas agora aplicados a Log Loss:

Gradiente em relação aos pesos ($w$):

partimos da Regra da Cadeia:
$$\frac{\partial J}{\partial w} = \frac{\partial J}{\partial \hat{y}} \cdot \frac{\partial \hat{y}}{\partial z} \cdot \frac{\partial z}{\partial w}$$

a derivação do custo em relação a predição fica:
$$\frac{\partial J}{\partial \hat{y}} = - \left[ \frac{y}{\hat{y}} - \frac{1-y}{1-\hat{y}} \right]= \frac{\hat{y} - y}{\hat{y}(1-\hat{y})}$$

a derivação da sigmóide fica:
$$\frac{\partial \hat{y}}{\partial z} = \sigma(z) \cdot (1 - \sigma(z)) = \hat{y}(1-\hat{y})$$

a derivação do termo z = $(\omega x +\beta)$ contraido a nível de simplificação:
$$\frac{\partial z}{\partial w} = x$$

unindo essas etapas anteriores, teremos:
$$\frac{\partial J}{\partial w} = \underbrace{\frac{\hat{y} - y}{\hat{y}(1-\hat{y})}}_{\text{etapa A}} \cdot \underbrace{\hat{y}(1-\hat{y})}_{\text{etapa B}} \cdot \underbrace{x}_{\text{etapa C}}$$

simplificando, teremos:
$$\frac{\partial J}{\partial w} = (\hat{y} - y)x$$

ou seja, na prática, temos que:
$$\frac{\partial J}{\partial w} = \frac{1}{N} \sum_{i=1}^{N} (\hat{y}_i - y_i) x_i$$

Gradiente em relação ao bias ($b$):
$$\frac{\partial J}{\partial b} = \frac{1}{N} \sum_{i=1}^{N} (\hat{y}_i - y_i)$$

As estratégias de atualização dos termos $\omega$ e $\beta$, seguem a mesma estrutura da Regressão Linear:
- Batch Gradient Descent: Processa o dataset inteiro para uma única atualização. Muito preciso para a direção do gradiente, mas custoso em memória.
- Stochastic Gradient Descent (SGD): Atualiza os pesos após cada exemplo. Introduz ruído que pode ajudar a escapar de platôs, mas a convergência é instável.
- Mini-Batch Gradient Descent: Processa pequenos subconjuntos (ex: 32, 64, 128 amostras). É a técnica mais utilizada, pois combina a eficiência com a estabilidade do gradiente.

Além dos métodos e do Learning Rate, a RL depende de outro hyperparâmetro, conhecido por "threshold", que é a fronteira de decisão, aqui é decidido a partir de que probabilidade calculada a amostra pode ser considerada da classe de ocorrência. Esse hiperparâmetro é muito importante dependendo do cenário que seu modelo está sendo aplicado, impactando diretamente em  métricas de Precisão, Recall e Acurácia por exemplo.

What is Logistic Regression?

Logistic Regression (LR) is a supervised machine learning model for binary classification. The goal is to find the relationship between a categorical target and its exploratory variables (features). Despite being called regression, RL outputs a binary categorical variable (e.g., 0/1, True/False, Yes/No, Female/Male), which is chosen based on the probability of an observation occurring. If this probability reaches a desired minimum limit, we classify the event as 1; otherwise, 0. The equation that shapes the model fit is given by:

Sigmoid Function:

$$f(x) = \frac{1}{1+e^{-\omega x +\beta}}$$

Thus, if we set our limit at 50%, then:
- If $f_\omega(x) \geq 0.5$, the prediction is class 1.
- If $f_\omega(x) < 0.5$, the prediction is class 0.

How does the RL model work?

Using Mean Squared Error (MSE) in this approach would generate a non-convex surface, greatly hindering the model's learning process. Due to the configuration of the function in use, the loss equation (LF) in this case is the Binary Cross Entropy (Log Loss), designated by:

$$J(w,b) = \frac{1}{N}\sum_{i=1}^{N}[y_i log(\hat{y}_i)+(1-y_i)log(1-\hat{y}_i)]$$

Where:

$y_i$: is the actual label (0 or 1).

$\hat{y}_i$: is the probability predicted by the model.

But where did this loss equation come from?

Let's go back a few steps and introduce a metric, 'log-odds', which is simply the chance of an event occurring, divided by the chance of it not occurring:

Log-Odds:

$$odds = \frac{P(event)}{1-P(event)}$$

Another important concept in probability is conditional probability, which is the chance of A happening, given that B has happened. This structure is defined by:

Conditional Probability:

$$Pr(y=1|X) = p(X)$$

where we calculate the probability of class 1 occurring given that event X has occurred. This function $p(X)$ is given by the Sigmoid Function presented earlier.

$$p(X) = \frac{1}{1+e^{-\omega x +\beta}}$$

Rearranging, we can write:

$$\ln\left(\frac{p(X)}{1-p(X)}\right) = \omega x + \beta$$

To arrive at the free-loop method, we will use the Maximum Likelihood (MLE) approach. For a binary event to occur, we use the Bernulli distribution, where the probability of an event y occurring is:

$$P(y|x) = \hat{y}^y (1-\hat{y})^{(1-y)}$$

where:
- $\hat{y}$ is the probability of the event occurring
- $y$ is the occurrence of the event
- $1-\hat{y}$ is the probability of the event not occurring
- $1-y$ is the non-occurrence of the event

In our entire dataset, the goal is to maximize the product of these probabilities (Likelihood). To transform this sequential product into a sequential sum, we apply the logarithm and invert the sign, going from a maximization to a minimization, and thus we arrive at the Log Loss:

$$J(w,b) = -\frac{1}{N} \sum_{i=1}^{N} \underbrace{\left[ y_i \ln(\hat{y}_i) + (1-y_i) \ln(1-\hat{y}_i) \right]}_{\text{Binary Cross Entropy}}$$

such that:
- If $y=1$: The cost is $-\ln(\hat{y})$. If $\hat{y} \to 1$, the cost is $0$. If $\hat{y} \to 0$, the cost tends to infinity.

- If $y=0$: The cost is $-\ln(1-\hat{y})$. If $\hat{y} \to 0$, the cost is $0$. If $\hat{y} \to 1$, the cost tends to infinity.

But how does RL learn?

Just like in Linear Regression, in RL there is an optimization process for updating the values ​​of $\omega$ and $\beta$ through Gradient Descent. This is the derivative with respect to the terms $\omega$ and $\beta$, but now applied to Log Loss:

Gradient with respect to the weights ($w$):

We start from the Chain Rule:

$$\frac{\partial J}{\partial w} = \frac{\partial J}{\partial \hat{y}} \cdot \frac{\partial \hat{y}}{\partial z} \cdot \frac{\partial z}{\partial w}$$

The derivation of the cost with respect to prediction is:
$$\frac{\partial J}{\partial \hat{y}} = - \left[ \frac{y}{\hat{y}} - \frac{1-y}{1-\hat{y}} \right]= \frac{\hat{y} - y}{\hat{y}(1-\hat{y})}$$

The derivation of the sigmoid is:

$$\frac{\partial \hat{y}}{\partial z} = \sigma(z) \cdot (1 - \sigma(z)) = \hat{y}(1-\hat{y})$$

The derivation of the term z = $(\omega x +\beta)$ contracted to the simplification level:

$$\frac{\partial z}{\partial w} = x$$

Combining these previous steps, we will have:

$$\frac{\partial J}{\partial w} = \underbrace{\frac{\hat{y} - y}{\hat{y}(1-\hat{y})}}_{\text{step A}} \cdot \underbrace{\hat{y}(1-\hat{y})}_{\text{stage B}} \cdot \underbrace{x}_{\text{stage C}}$$

Simplifying, we have:

$$\frac{\partial J}{\partial w} = (\hat{y} - y)x$$

In other words, in practice, we have:

$$\frac{\partial J}{\partial w} = \frac{1}{N} \sum_{i=1}^{N} (\hat{y}_i - y_i) x_i$$

Gradient with respect to the bias ($b$):

$$\frac{\partial J}{\partial b} = \frac{1}{N} \sum_{i=1}^{N} (\hat{y}_i - y_i)$$

The strategies for updating the terms $\omega$ and $\beta$ follow the same structure as Linear Regression:

- Batch Gradient Descent: Processes the entire dataset for a single update. Very accurate for gradient direction, but memory expensive.
- Stochastic Gradient Descent (SGD): Updates the weights after each example. Introduces noise that can help to escape plateaus, but convergence is unstable.
- Mini-Batch Gradient Descent: Processes small subsets (e.g., 32, 64, 128 samples). It is the most widely used technique because it combines efficiency with gradient stability.

In addition to the methods and the Learning Rate, RL depends on another hyperparameter, known as the "threshold," which is the decision boundary. Here, it is decided from what calculated probability the sample can be considered to belong to the occurrence class. This hyperparameter is very important depending on the scenario in which your model is being applied, directly impacting metrics such as Precision, Recall, and Accuracy.
